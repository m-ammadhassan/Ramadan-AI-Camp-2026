```python
#Question No.7
subjectEnglish = int(input("Enter your English subject Obtained Marks: "))
subjectUrdu = int(input("Enter your Urdu subject Obtained Marks: "))
subjectMathematics = int(input("Enter your Mathematics subject Obtained Marks: "))
subjectScience = int(input("Enter your Science subject Obtained Marks: "))
subjectIslamiat = int(input("Enter your Islamiat subject Obtained Marks: "))

totalObtainedMarks = subjectEnglish + subjectUrdu + subjectMathematics + subjectScience + subjectIslamiat

percentage = (totalObtainedMarks / 500) * 100

if percentage >= 90:
    gradeObtained = "A*"
elif percentage >= 80 and percentage < 90:
    gradeObtained = "A"
elif percentage >= 70 and percentage < 80:
    gradeObtained = "B"
elif percentage >= 60 and percentage < 70:
    gradeObtained = "C"
elif percentage >= 50 and percentage < 60:
    gradeObtained = "D"
elif percentage >= 40 and percentage < 50:
    gradeObtained = "E"
else:
    gradeObtained = "F"

print("\n||===Marksheet===||")
print("English", subjectEnglish, "out of 100.", "\nUrdu", subjectUrdu, "out of 100.", "\nMathematics", subjectMathematics, "out of 100.", "\nScience", subjectScience, "out of 100.", "\nIslamiat", subjectIslamiat, "out of 100.")
print("Total Marks:", 500, "\nObtained Marks:", totalObtainedMarks, "\nPercentage:", percentage, "\nGrade:", gradeObtained)
```
