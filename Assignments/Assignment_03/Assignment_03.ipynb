```python
#Question No.1
print("||===Calculator===||")
firstValue = float(input("Enter the First Number: "))
secondValue = float(input("Enter the Second Number: "))
print("\nThe Calculator Operations Available")
print("1 - Addition", "\n2 - Subtraction", "\n3 - Multiplication", "\n4 - Division", "\n5 - Power")
option = int(input("\nSelect from 1 to 5: "))
if option == 1:
    print("Result = ", firstValue + secondValue)
elif option == 2:
    print("Result = ", firstValue - secondValue)
elif option == 3:
    print("Result = ", firstValue * secondValue)
elif option == 4:
    print("Result = ", firstValue / secondValue)
elif option == 5:
    print("Result = ", firstValue ** secondValue)
else:
    print("ERROR: Invalid Choice!")
```

```python
#Question No.2
myList = ["Red", "Yellow", 28, "Green", 22, 40, "White", 32, "Blue", 24, 60, 18, "Purple", "Pink"]
isNumericValue = False
for num in myList:
    if type(num) == int or type(num) == float:
        isNumericValue = True

if isNumericValue:
    print("Yes, there is a numeric value in a List")
else:
    print("No, there is no any numeric value in a List")
```

```python
#Question No.5
marksList = [90, 70, 95, 85, 90, 60, 80, 85]
duplicateValuesList = []
for value in marksList:
    if marksList.count(value) > 1 and value not in duplicateValuesList:
        duplicateValuesList.append(value)
print("The duplicate values list:", duplicateValuesList)
```
